```python
import random
```

```python
def transfer_bitcoin(sender, receiver, amount, users):
    print("\nStep 1: Validating input details")

    # Validation
    if not sender or not receiver or amount <= 0:
        print("Invalid transaction details!\n")
        return
    
    if sender not in users:
        print("Sender not found!\n")
        return

    if receiver not in users:
        print("Receiver not found!\n")
        return
    
    if users[sender]["balance"] < amount:
        print("Insufficient balance!\n")
        return

    

    print(f"\nSender: {users[sender]['name']} ({sender})")
    print(f"Receiver: {users[receiver]['name']} ({receiver})")
    print(f"Amount: {amount} BTC\n")

    # Transaction
    transaction = {
        "sender": sender,
        "receiver": receiver,
        "amount": amount,
        "transaction_id": random.randint(100000, 999999)
    }

    # Update balances
    users[sender]["balance"] -= amount
    users[receiver]["balance"] += amount

    print(f"Transaction ID: {transaction['transaction_id']}")
    print("Transaction successful!\n")

    return transaction, users
```

```python
print("Sanya Khare")
print("TCA237045")
if __name__ == "__main__":
    users = {
        "192.168.1.2": {"name": "Aditya", "balance": 80.0},
        "192.168.1.3": {"name": "Akshat", "balance": 20.0},
        "192.168.1.4": {"name": "Manish", "balance": 30.0},
        "192.168.1.5": {"name": "Aashish", "balance": 50.0}
    }

    sender = input("Enter sender IP: ")
    receiver = input("Enter receiver IP: ")
    amount = float(input("Enter amount: "))

    result = transfer_bitcoin(sender, receiver, amount, users)

    if result:
        transaction, updated_balances = result

        print("Transaction Details:")
        print(transaction)
        

        print("\nUpdated Balances:")
        for user, balance in updated_balances.items():
            print(f"{user}: {balance}")
```
