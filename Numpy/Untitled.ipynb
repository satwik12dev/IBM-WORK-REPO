```python

import numpy as np

data = np.genfromtxt("train_and_test2.csv", delimiter=",", skip_header=1)
```

```python
null_mask = np.isnan(data)

print("Null values present:", np.any(null_mask))
print("Number of null values:", np.sum(null_mask))
```

```python
# Keep only rows without any NaN
data_no_null = data[~np.isnan(data).any(axis=1)]

print("Shape after removing null rows:", data_no_null.shape)
```

```python
Q1 = np.percentile(data_no_null, 25, axis=0)
Q3 = np.percentile(data_no_null, 75, axis=0)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
# Keep rows where all values are within bounds
outlier_mask = np.all(
    (data_no_null >= lower_bound) & (data_no_null <= upper_bound),
    axis=1
)

clean_data = data_no_null[outlier_mask]

print("Shape after removing outliers:", clean_data.shape)
```
